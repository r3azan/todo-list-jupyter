```python
ToDo_list=[]
```

```python
def add():
    x = input("Enter task to add: ")
    ToDo_list.append(x)
    print(f'"{x}" added successfully.')
```

```python
def remove():
    x = input('REMOVE: ')
    if x in ToDo_list:
        ToDo_list.remove(x)
        print(f'"{x}" removed successfully.')
    else:
        print('Task not found.')
```

```python
def count():
    print(f'You have {len(ToDo_list)} task(s).')
```

```python
def view():
    print('\nMY TO DO LIST:')
    if not ToDo_list:
        print('No tasks yet.')
    else:
        for i, task in enumerate(ToDo_list, start=1):
            print(f'{i}. {task}')
```

```python
while True:
    try:
        x = int(input('''
Enter your choice:
1- Add task
2- Remove task
3- Count tasks
4- View tasks
5- Exit
'''))

        if x == 1:
            add()
        elif x == 2:
            remove()
        elif x == 3:
            count()
        elif x == 4:
            view()
        elif x == 5:
            print("Goodbye!")
            break
        else:
            print('Invalid input, try again!')

    except ValueError:
        print("Please enter a valid number.")
```
